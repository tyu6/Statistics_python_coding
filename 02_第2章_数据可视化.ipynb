### 代码框1-23 matplotlib绘图的一些基本操作

```python
#导入相关模块
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei'                              #显示中文
plt.rcParams['axes.unicode_minus'] = False                              #显示负号
#生成绘图数据
np.random.seed(100)                                                     #设置随机种子数
x = np.random.standard_normal(200)                                      #产生200个标准正态分布随机数，赋值给x
y = 1+2*x+np.random.normal(0,1,200)                                     #产生变量y的随机数
#绘制图像内容
plt.figure(figsize = (8,6))                                             #设置图形大小
plt.scatter(x,y,marker = 'o',c='none',edgecolors =['black'] )           #绘制散点图，改进代码c='none','black'添加方括弧['black']
# pf = np.polyfit(x, y, deg=1)
# pf
# fit = np.polyld([pf])
# # y_hat = fit[x]
# # plt.plot(x, y_hat, c='b')
plt.plot(x.mean(),y.mean(),'ro',markersize=20,fillstyle='bottom')       #添加均值点并设置点的大小，颜色和填充类型
plt.axhline(y.mean(),color='black',ls='--',lw=1)                        #添加均值水平线并设置线的颜色，线型，线宽
plt.axvline(x.mean(),color='black',ls = (0,(5,10)),lw=1)                #增加均值垂直线
#绘制其他组件
plt.grid(linestyle=':')                                                 #添加网格线
ax = plt.gca()                                                          #得到当前操作图像
ax.spines['right'].set_color('#4169E1')                                 #设置边框颜色
ax.spines['left'].set_color('#4169E1')
ax.spines['top'].set_color('royalblue')
ax.spines['bottom'].set_color('b')
plt.text(x=0.4, y=-3, s=r'$\hat{y}=\hat{\beta}_0+\hat{\beta}_1x$',fontdict={'size':12,'bbox':{'fc':'pink','boxstyle':'round'}},ha='left')
                                                                        #添加文本注释
#plt.annotate(x=-0.6, y=3, text = r'均值点',xy = (x.mean(),y.mean()),arrowprops = {'headwidth':10,'headlenth':5,'width':4,'facecolor':'r','shrink':0.1,},fontsize=14, color='red',ha='right')
plt.title('散点图及拟合直线\n并为图形增加新的元素', fontsize=14,color='white')   #添加标题
plt.xlabel('x=自变量', fontsize=10,color='white')                                #添加x轴标题
plt.ylabel('y=因变量', fontsize=10,color='white')                                #添加y轴标题
plt.legend(['拟合直线'], loc='best')                                             #添加图例
plt.show()
```

### 代码框1-25 画布的不等分布局

```python
#使用GridSpec函数布局
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6,5))                       #设置图形大小
grid=plt.GridSpec(3,3)                                #生成3*3网格
plt.subplot(grid[0,:2])                               #占据第一行，前两列
plt.subplot(grid[0,2])                                #占据第一行，第三列
plt.subplot(grid[1,:1])                               #占据第二行，第一列
plt.subplot(grid[1,1:])                               #占据第二行，后两列
plt.subplot(grid[2,:3])                               #占据第三行前三列
fig.tight_layout                                      #紧凑布局
```

```python
#使用subplot2grid函数布局
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6, 4.5))
ax1 = plt.subplot2grid((3,3),(0,0),rowspan=2,colspan=2)                #第一行第一幅图，占据两行两列
ax2 = plt.subplot2grid((3,3),(0,2),rowspan=3,colspan=1)                #第一行第二幅图，占据三行一列
ax3 = plt.subplot2grid((3,3),(2,0),rowspan=1,colspan=1)                #第三行第一幅图，占据一行一列
ax4 = plt.subplot2grid((3,3),(2,1),rowspan=1,colspan=1)                #第三行第二幅图，占据一行一列
fig.tight_layout()
```

```python
#使用add_gridspec函数布局
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(1010)
fig = plt.figure(figsize=(7,4))
spec = fig.add_gridspec(nrows=2, ncols=6, width_ratios=[1,1,1,1,1,1],height_ratios=[1,2])
fig.add_subplot(spec[0,1:3])                        #占据第1行的2到3列
fig.add_subplot(spec[0,3:])                         #占据第1行的后三列
ax = fig.add_subplot(spec[:,0])                     #占据第1列的所有行
ax = fig.add_subplot(spec[1,1:4])                   #占据第2行的2-4列
ax = fig.add_subplot(spec[1,4:])                    #占据第2行第4列后面的所有列
fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.sans-serif'] = ['SimHei']
# plt.rcParams['font.sans-serif'] = ['SimHei']
# plt.rcParams['axes.unicode_minus'] = False
#plt.rcParams['patch.force_edgecolor'] = True
df = pd.read_csv('F:\JupyterData\Statistics_Pythoncoding\data\chap02\example2_2.csv', encoding = 'gbk')

#创建角度
attributes = list(df.columns[1:])
values  = list(df.values[:,1:])
names = list(df.values[:,0])
angles = [n / float(len(attributes)) * 2 * np.pi for n in range(len(attributes))]

#关闭图画
angles += angles[:1]
values = np.asarray(values)
values = np.concatenate([values, values[:,0:1]], axis=1)

#创建图形
sns.set_style('darkgrid')
plt.figure(figsize=(8,8))
for i in range(4):
    ax = plt.subplot(2,2,i+1,polar=True)
    ax.plot(angles,values[i])
    ax.set_yticks(np.arange(500,16000,3500))
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(attributes)
    ax.set_title(names[i],fontsize=12,color='red')
    plt.tight_layout()
plt.show()
    
    
```

### 代码框2-18 绘制雷达图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
df = pd.read_csv('F:\JupyterData\Statistics_Pythoncoding\data\chap02\example2_2.csv', encoding = 'gbk')
labels = np.array(df['支出项目'])
datalenth=8

df1=np.array(df['北京'])
df2=np.array(df['天津'])
df3=np.array(df['上海'])
df4=np.array(df['重庆'])

angles = np.linspace(0,2*np.pi,datalenth,endpoint=False)
df1=np.concatenate((df1,[df1[0]]))
df2=np.concatenate((df2,[df2[0]]))
df3=np.concatenate((df3,[df3[0]]))
df4=np.concatenate((df4,[df4[0]]))
angles = np.concatenate((angles,[angles[0]]))

plt.figure(figsize=(6,6),facecolor='lightyellow')
plt.polar(angles,df1,'r--',linewidth=1,label='北京')
plt.polar(angles,df2,'b',linewidth=1,label='天津')
plt.polar(angles,df3,'k',linewidth=1,label='上海')
plt.polar(angles,df4,'g',linewidth=1,label='重庆')
plt.thetagrids(range(0,360,45),labels)
plt.grid(linestyle="-",linewidth=0.5,color='gray',alpha=0.5)
plt.legend(loc='upper right',bbox_to_anchor=(1.1,1.1))
plt.show()
```
